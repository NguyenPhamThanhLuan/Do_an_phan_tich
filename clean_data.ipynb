```python
# 1.điều chỉnh lại cột bắt đàu và nghỉ
df['StartDate'] = pd.to_datetime(df['StartDate'])
df['ExitDate'] = pd.to_datetime(df['ExitDate'])
```

```python
# xem lại coi sau khi chuyển 
df[['StartDate','ExitDate']]
```

```python
# Kiểm tra ExitDate sớm hơn StartDate
invalid_dates = df[df['ExitDate'] < df['StartDate']]
print(invalid_dates)
# Kiểm tra nhân viên chưa có ExitDate 
active_employees = df[df['ExitDate'].isna()]
print(f"Số lượng nhân viên vẫn đang làm việc: {active_employees.shape[0]}")
```

```python
# Chuyển cột ExitDate về kiểu chuỗi (string)
df['ExitDate'] = df['ExitDate'].dt.strftime('%Y-%m-%d')
```

```python
df['ExitDate'].info()
```

```python

df['ExitDate'] = df['ExitDate'].fillna('yes')# xử lí null trong cột exit date
```

```python

df['ExitDate'].head(10)
```

```python
#  EmpID,ADEmail coi có trùng k bởi vi người chỉ có 1 mã nv và 1 email
check_dup = df[df.duplicated(subset=['EmpID', 'FirstName', 'LastName', 'ADEmail'], keep=False)]
print(check_dup)
```

```python
df[['BusinessUnit','EmployeeStatus','EmployeeType','ExitDate']].head(10) 
```

```python

# Nếu "ExitDate" là null hoặc "Yes", trạng thái nhân viên sẽ là "Active"
df.loc[(df['ExitDate'] == 'yes'), 'EmployeeStatus'] = 'Active'
# Nếu "ExitDate" có giá trị và khác "Yes", trạng thái nhân viên sẽ là "Off"
df.loc[(df['ExitDate'] != 'yes'), 'EmployeeStatus'] = 'Off'
```
