```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('employ_clean.xlsx')
```

```python
df.isnull().sum()
```

1

```python
#1V Bar chart  phn tch s lng nhn vin theo chc v, vi mu sc khc v cc chc v  di
plt.figure(figsize=(12, 6))
sns.countplot(y='Title', data=df, order=df['Title'].value_counts().index, palette='Set1')
plt.title('s lng nv theo cng vic')
plt.xlabel('s lng NV')
plt.ylabel('vic lm')
plt.show()
```

2

```python
#2 Pie chart  phn tch s lng nhn vin theo gii tnh
gender_counts = df['GenderCode'].value_counts()
plt.figure(figsize=(8, 8))
gender_counts.plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3')
# Thm ch thch
plt.title('Distribution of Employees by Gender')
plt.ylabel('')
plt.legend(labels=['Nam (1)', 'N (0)'], loc='best')  # Ch thch thm
plt.show()

```

```python
# m bo StartDate l kiu datetime
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')

# To ct 'Year' t ct StartDate
df['Year'] = df['StartDate'].dt.year# .
```

```python
# m bo StartDate l kiu datetime
df[['Year','StartDate']]# thy . year khng l ch ly nm thoi ( dt l datetime)
```

```python
new_hires_by_year = df.groupby('Year').size()
new_hires_by_year 
```

```python
# V biu  ng s lng nhn vin mi theo nm
plt.plot(new_hires_by_year.index, new_hires_by_year.values, marker='o', color='yellow') # xem inddex vs xem value
plt.xlabel('Nm')
plt.ylabel('S lng nhn vin mi')
plt.title('S lng nhn vin mi theo nm')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the bar chart for number of employees by age group and employment status
age_groups = pd.cut(df['Age'], bins=[0, 30, 40, 50, 60, 100], labels=['0-30', '31-40', '41-50', '51-60', '60+'])
grouped_data = df.groupby([age_groups, 'EmployeeStatus']).size().unstack()#  y cng nhm li  nu c c trn th hiu thoi  un n ang seri thid i sang data frsm
colors = ['#2ca02c', '#9467bd']
# Plot # ly kq r v thoi
grouped_data.plot(kind='bar', stacked=True, figsize=(10,6),color = colors )
plt.title('s lng nhn vin cn lm hay  ngh')
plt.xlabel(' tui')
plt.ylabel('s lng nhn vin')
plt.xticks(rotation=0)
plt.legend(title='trng thi nhn vin', bbox_to_anchor=(1.2, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
