```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('employ_clean.xlsx')
```

```python
df.isnull().sum()
```

1

```python
#1Vẽ Bar chart để phân tích số lượng nhân viên theo chức vụ, với màu sắc khác và các chức vụ ở dưới
plt.figure(figsize=(12, 6))
sns.countplot(y='Title', data=df, order=df['Title'].value_counts().index, palette='Set1')
plt.title('số lương nv theo công việc')
plt.xlabel('số lượng NV')
plt.ylabel('việc làm')
plt.show()
```

2

```python
#2 Pie chart để phân tích số lượng nhân viên theo giới tính
gender_counts = df['GenderCode'].value_counts()
plt.figure(figsize=(8, 8))
gender_counts.plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3')
# Thêm chú thích
plt.title('Distribution of Employees by Gender')
plt.ylabel('')
plt.legend(labels=['Nam (1)', 'Nữ (0)'], loc='best')  # Chú thích thêm
plt.show()

```

```python
# Đảm bảo StartDate là kiểu datetime
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')

# Tạo cột 'Year' từ cột StartDate
df['Year'] = df['StartDate'].dt.year# .
```

```python
# Đảm bảo StartDate là kiểu datetime
df[['Year','StartDate']]# thấy . year không là chỉ lấy năm thoi ( dt là datetime)
```

```python
new_hires_by_year = df.groupby('Year').size()
new_hires_by_year 
```

```python
# Vẽ biểu đồ đường số lượng nhân viên mới theo năm
plt.plot(new_hires_by_year.index, new_hires_by_year.values, marker='o', color='yellow') # xem inddex vs xem value
plt.xlabel('Năm')
plt.ylabel('Số lượng nhân viên mới')
plt.title('Số lượng nhân viên mới theo năm')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the bar chart for number of employees by age group and employment status
age_groups = pd.cut(df['Age'], bins=[0, 30, 40, 50, 60, 100], labels=['0-30', '31-40', '41-50', '51-60', '60+'])
grouped_data = df.groupby([age_groups, 'EmployeeStatus']).size().unstack()# ở đây cũng nhóm lại đó nếu có đọc trên thì hiẻu thoi  un nó đang seri thid đổi sang data frsm
colors = ['#2ca02c', '#9467bd']
# Plot # láy kq r vẽ thoi
grouped_data.plot(kind='bar', stacked=True, figsize=(10,6),color = colors )
plt.title('số lượng nhân viên còn làm hay đã nghỉ')
plt.xlabel('độ tuổi')
plt.ylabel('só lượng nhân viên')
plt.xticks(rotation=0)
plt.legend(title='trạng thái nhân viên', bbox_to_anchor=(1.2, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
